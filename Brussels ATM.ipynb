```python
import pandas as pd
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
path = '/home/ahsan/Dropbox/data/ATM in Brussels/atms.csv'
atms = pd.read_csv(path, sep= ';')
print(atms.shape)
atms[['ATM (adress in French)','Coordinates_X','Coordinates_Y']].head()
```

```python
# make sure the value of resolution is a lowercase L,
#  for 'low', not a numeral 1

world_map = Basemap(projection='ortho', lat_0=20, lon_0=70,
              resolution = 'l', area_thresh = 500.0)
fig = plt.figure(figsize=(10,10))
# world_map.drawcoastlines()
# world_map.drawcountries()
world_map.fillcontinents(color='coral')
world_map.drawmapboundary()
plt.show()
```

```python
europe_map = Basemap(projection='robin', lat_0=20, lon_0=20,
              resolution = 'l', area_thresh = 1000.0)
fig = plt.figure(figsize=(15,15))
# europe_map.drawcoastlines()
europe_map.drawcountries()
europe_map.fillcontinents(color='coral')
europe_map.drawmapboundary()
plt.show()
```

```python
belgium_map = Basemap(width=12000000,height=9000000,projection='merc', lat_0=57, lon_0=-135,
    resolution = 'l', area_thresh = 1000.0,
       llcrnrlon=1, llcrnrlat=49,
    urcrnrlon=8, urcrnrlat=53)
fig = plt.figure(figsize=(10,10))
belgium_map.drawcoastlines()
belgium_map.drawcountries()
belgium_map.fillcontinents(color='coral')
belgium_map.drawmapboundary()
 
#belgium_map.drawmeridians(np.arange(0, 360, 30))
#belgium_map.drawparallels(np.arange(-90, 90, 30))

def mapLocation():
    
    for i in range(atms.shape[0]):
        if ~np.isnan(atms['Coordinates_X'][i]):
            x,y = belgium_map(atms['Coordinates_Y'][i], atms['Coordinates_X'][i])
            belgium_map.plot(x, y, 'bo', markersize=3)

mapLocation()
 
plt.show()
```
